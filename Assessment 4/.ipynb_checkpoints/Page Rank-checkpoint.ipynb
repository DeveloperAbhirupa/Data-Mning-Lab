```python
import networkx as nx

G=nx.Graph()
G.add_nodes_from(['A','B','C','D','E','F'])
print(G.nodes.data())
```

```python
G.add_edges_from([('A','B'),('A','C'),('B','A'),('B','D'),('B','E'),('B','F'),('C','A'),('C','F'),('F','E')])
print(G.edges.data())
dg=nx.DiGraph()
```
