```python
import networkx as nx

dg=nx.DiGraph()
dg.add_nodes_from(['A','B','C','D','E','F'])
# print(G.nodes.data())
```

```python

dg.add_edges_from([('A','B'),('A','C'),('B','A'),('B','D'),('B','E'),('B','F'),('C','A'),('C','F'),('F','E')])
# print(G.edges.data())

nx.draw(dg, with_labels=True)
```

```python
# pg=nx.pagerank(dg, alpha=0.85, personalization=None, max_iter=8 dangling=None)
pr=nx.pagerank(dg,alpha=0.85, max_iter=100)
print(pr)
```

```python
dg1=nx.DiGraph()
dg1.add_nodes_from([1,2,3,4,5,6,7,8,9])
dg1.add_edges_from([(1,2),(1,3),(2,4),(3,9),(4,5),(4,7),(5,7),(6,7),(7,8),(7,9),(8,9)])
nx.draw(dg1, with_labels=True)
```

```python
a,h=nx.hits(dg1, max_iter=100, tol=1e-08, nstart=None, normalized=True)
print("Authority Score:", a)
print("\n\nHub Score:", h)
```
